```python
#Assignment 5 Nimmagadda Shree Deepthi 2303A52303 Batch 37
#Read the data with pandas and find features and target variables
#Find the target variable
#Plot a graph between features and target
#Normalize the give data[0 to 1]
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler


df = pd.read_csv('Salary_Data.csv')

features = df[['YearsExperience']]
target = df['Salary']
print("Target Variable:", target.name)

plt.scatter(features, target, color='blue')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

scaler = MinMaxScaler()

features_normalized = scaler.fit_transform(features)

target_normalized = scaler.fit_transform(target.values.reshape(-1, 1))

df_normalized = pd.DataFrame({
    'YearsExperience': features_normalized.flatten(),
    'Salary': target_normalized.flatten()
})

print(df_normalized.head())
```

```python
#Assignment 5 Nimmagadda Shree Deepthi 2303A52303 Batch 37
#Read the data with pandas and find features and target variables
#Plot a graph between features and target
#Normalize the give data[0 to 1]
#Convert string valued features to number
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

df = pd.read_csv('diabetes_data_upload.csv')


label_encoders = {}
for column in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le


features = df.drop(columns=['class'])

plt.scatter(df['Age'], target, color='blue')
plt.title('Age vs Class')
plt.xlabel('Age')
plt.ylabel('Class')
plt.show()


scaler = MinMaxScaler()
features_normalized = scaler.fit_transform(features)

df_normalized = pd.DataFrame(features_normalized, columns=features.columns)

df_normalized[target.name] = target.values

print(df_normalized.head())


```
